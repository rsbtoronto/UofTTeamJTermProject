## Column Heading change
#### Column heading were long and sentence like. We created a metadata file which would check for strings in the spreadsheet and assign the column name dynamically. Code would check for Text1 and Text2 in the header and assign the column name from corresponding row in column directory. 

```python
def ColumnNameChange(i):    
    for x in range(len(mdf.index)):
        c1 = mdf[mdf.index == x].Text1.to_string(index=False)       
        c2 = mdf[mdf.index == x].Text2.to_string(index=False)  
        c3 = mdf[mdf.index == x].Cols.to_string(index=False)
        if (i.find(c1) != -1) & (i.find(c2) != -1):
            df.rename(columns = {i:c3}, inplace=True)            
            return "changed"
        
for a in listdf:
    ColumnNameChange(a)
    

df.head(5) 
```

## Handling Missing data
#### Some countries had data from 195, but missing several years in recent years, some were missing both at the beginning period and ending period. For e.g. Cuba. 
#### We used forward fill method to populate the data for future years and followed by backward fill to populate the beginning years as our first country in the set ‘Albania’ didn’t have data for first few years. We encountered a problem with that approach as the data of one country used to prefill with another. We added the condition to check for the country and make sure we apply fillna to same country for the given set of years. We looped thru each country one by one to apply this function.

```python
sdf = df

a = "India" 
for x in df.Country:
    if a != x:
        sdf[sdf.Country == x] = df[df.Country == x].fillna(method='ffill')       
        a = x
    #print(x)
for x in df.Country:
    if a != x:
        sdf[sdf.Country == x] = df[df.Country == x].fillna(method='bfill')       
        a = x

sdf.head(5)
```

## Creating the data file with per 100K population

```python
dfpct = df
dfpct.head(5)
```

```python
per100K = 100000
dfpct.PopMale = dfpct.PopMale / dfpct.TotalPop * per100K
dfpct.PopFemale = dfpct.PopFemale / dfpct.TotalPop * per100K
TotalDeath = dfpct.TransportAccAll  + dfpct.FallsAll +  dfpct.DrowningAll + dfpct.FireAll + dfpct.PoisonAll + dfpct.SuicideAll + dfpct.AssaultAll + dfpct.OtherAll

dfpct.TransportAccAll = dfpct.TransportAccAll / dfpct.TotalPop * per100K
dfpct.TransportAccMale = dfpct.TransportAccMale / dfpct.TotalPop * per100K
dfpct.TransportAccFemale = dfpct.TransportAccFemale / dfpct.TotalPop * per100K
dfpct.FallsAll = dfpct.FallsAll / dfpct.TotalPop * per100K
dfpct.FallsMale = dfpct.FallsMale / dfpct.TotalPop * per100K
dfpct.FallsFemale = dfpct.FallsFemale / dfpct.TotalPop * per100K
dfpct.FallsAll = dfpct.FallsAll / dfpct.TotalPop * per100K
dfpct.FallsMale = dfpct.FallsMale / dfpct.TotalPop * per100K
dfpct.FallsFemale = dfpct.FallsFemale / dfpct.TotalPop * per100K
dfpct.DrowningAll = dfpct.DrowningAll / dfpct.TotalPop * per100K
dfpct.DrowningMale = dfpct.DrowningMale / dfpct.TotalPop * per100K
dfpct.DrowningFemale = dfpct.DrowningFemale / dfpct.TotalPop * per100K
dfpct.FireAll = dfpct.FireAll / dfpct.TotalPop * per100K
dfpct.FireMale = dfpct.FireMale / dfpct.TotalPop * per100K
dfpct.FireFemale = dfpct.FireFemale/ dfpct.TotalPop * per100K
dfpct.PoisonAll = dfpct.PoisonAll / dfpct.TotalPop * per100K
dfpct.PoisonMale = dfpct.PoisonMale / dfpct.TotalPop * per100K
dfpct.PoisonFemale = dfpct.PoisonFemale / dfpct.TotalPop * per100K
dfpct.SuicideAll = dfpct.SuicideAll / dfpct.TotalPop * per100K
dfpct.SuicideMale = dfpct.SuicideMale / dfpct.TotalPop * per100K
dfpct.SuicideFemale = dfpct.SuicideFemale / dfpct.TotalPop * per100K
dfpct.AssaultAll = dfpct.AssaultAll / dfpct.TotalPop * per100K
dfpct.AssaultMale = dfpct.AssaultMale / dfpct.TotalPop * per100K
dfpct.AssaultFemale = dfpct.AssaultFemale / dfpct.TotalPop * per100K
dfpct.OtherAll = dfpct.OtherAll / dfpct.TotalPop * per100K
dfpct.OtherMale = dfpct.OtherMale / dfpct.TotalPop * per100K
dfpct.OtherFemale = dfpct.OtherFemale / dfpct.TotalPop * per100K
dfpct["TotalDeath"] = TotalDeath / dfpct.TotalPop * per100K 

dfpct.fillna(method='ffill', inplace=True)
dfpct.fillna(method='bfill', inplace=True)
```
